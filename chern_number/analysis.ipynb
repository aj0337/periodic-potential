```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
bnd_indices = [73,74,75]
output_dir = "data/berry_curvature_blg"

for bnd_idx in bnd_indices:
    # Load the data
    data = np.load(f'{output_dir}/berry_curvature_band_{bnd_idx}.npy')

    kx, ky, berry_curvature = data[:, 0], data[:, 1], data[:, 2] #multiply berry_curvature by dkx*dky to get scaled numbers

    # Create a heatmap
    plt.figure(figsize=(8, 6))
    plt.tricontourf(kx, ky, berry_curvature, levels=100, cmap='viridis')  # Adjust cmap for better contrast if needed
    plt.colorbar(label="Berry Curvature")
    plt.xlabel(r"$k_x$")
    plt.ylabel(r"$k_y$")
    plt.title("Berry Curvature Heatmap")
    plt.show()

```
