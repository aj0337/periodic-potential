```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

# Define the simple function
def func(x):
    return x**2

# Method 1: Using scipy nquad integration
def scipy_integration():
    result, error = integrate.nquad(func, [[0, 1]])
    return result

# Method 2: Summation over discrete points
def summation_integration(num_points=1000):
    x_vals = np.linspace(0, 1, num_points)
    dx = x_vals[1] - x_vals[0]  # Compute the step size
    y_vals = func(x_vals)
    area_sum = np.sum(y_vals * dx)
    return area_sum

# Compare the results
scipy_result = scipy_integration()
summation_result = summation_integration()

print(f"Result using scipy nquad integration: {scipy_result}")
print(f"Result using summation method: {summation_result}")

# Optionally plot the function and the rectangles used in the summation method
def plot_summation(num_points=1000):
    x_vals = np.linspace(0, 1, num_points)
    y_vals = func(x_vals)

    plt.plot(x_vals, y_vals, label='f(x) = x^2', color='blue')
    plt.fill_between(x_vals, y_vals, color='lightblue', alpha=0.6, label='Summation approximation')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Function and Area Under the Curve')
    plt.legend()
    plt.show()

# Plot the summation approximation
plot_summation()
```

```python
# Define the 2D function
def func(x, y):
    return x**2 + y**2

# Method 1: Using scipy nquad integration for 2D
def scipy_integration_2d():
    result, error = integrate.nquad(func, [[0, 1], [0, 1]])
    return result

# Method 2: Summation over discrete points for 2D
def summation_integration_2d(num_points=100):
    x_vals = np.linspace(0, 1, num_points)
    y_vals = np.linspace(0, 1, num_points)
    dx = x_vals[1] - x_vals[0]
    dy = y_vals[1] - y_vals[0]

    area_sum = 0
    for x in x_vals:
        for y in y_vals:
            area_sum += func(x, y) * dx * dy

    return area_sum

# Compare the results
scipy_result_2d = scipy_integration_2d()
summation_result_2d = summation_integration_2d()

print(f"Result using scipy nquad integration (2D): {scipy_result_2d}")
print(f"Result using summation method (2D): {summation_result_2d}")

# Optionally plot the 2D function (heatmap) and approximation
def plot_summation_2d(num_points=100):
    x_vals = np.linspace(0, 1, num_points)
    y_vals = np.linspace(0, 1, num_points)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = func(X, Y)

    plt.contourf(X, Y, Z, levels=50, cmap='viridis')
    plt.colorbar(label='f(x, y) = x^2 + y^2')
    plt.title('2D Function Contour Plot')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Plot the summation approximation
plot_summation_2d()
```
