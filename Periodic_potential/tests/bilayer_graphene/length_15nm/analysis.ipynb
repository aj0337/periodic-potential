```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def split_data_by_bands(filename):
    bands = []
    current_band_data = []

    with open(filename, 'r') as file:
        next(file)

        for line in file:
            if line.strip():  # Check if the line is not empty
                data = line.split()
                current_band_data.append([float(val) for val in data])
            else:
                bands.append(current_band_data)
                current_band_data = []

        if current_band_data:
            bands.append(current_band_data)

    return np.array(bands)
```

```python
potentials = ["0", ".025", ".045", ".050",".055",".090",".110"]#, ".325", ".350", ".375", ".400"]
# ("G  "    0.00000,"A  "    0.42422,"P  "    0.84847,"X  "    1.27272)
xtics =  [    0.00000,    0.85137,    1.27707,    1.29243,    2.56950,    2.58486,    3.86202]
xlabels = ["G","P","X","W","Y","G","W"]
for potential in potentials:

    bands = split_data_by_bands(f"sine/potential_{potential}/bulkek.dat")
    vb1 =  bands[139][:,4]
    cb1 = bands[140][:,4]
    klen = bands[139][:,3]
    plt.plot(klen,vb1)
    plt.plot(klen,cb1,label=f"V = {potential} eV")
    # plt.ylim(-0.015,0.015)
    # plt.xlim(0.83,0.87)
    plt.ylabel("E (eV)")
    plt.xlabel("k")
    plt.axhline(0, color='gray', linewidth=1.5)
    # plt.xticks(labels=xlabels,ticks=xtics)
    plt.legend()
    plt.show()
```
