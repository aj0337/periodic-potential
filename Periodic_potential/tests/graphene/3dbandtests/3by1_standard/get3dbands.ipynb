```python
import plotly.graph_objs as go
import numpy as np
```

```python
def split_data_by_bands(filename):
    bands = []
    current_band_data = []

    with open(filename, 'r') as file:

        next(file)

        for line in file:
            if line.strip():
                data = line.split()
                current_band_data.append([float(val) for val in data])
            else:
                bands.append(current_band_data)
                current_band_data = []


        if current_band_data:
            bands.append(current_band_data)

    return np.array(bands)
```

```python
filename = "potential_0/bulkek.dat"
bands = split_data_by_bands(filename)
b1 = bands[1]
b2 = bands[2]

fig = go.Figure()

fig.add_trace(go.Scatter3d(x=b1[:,0], y=b1[:,1], z=b1[:,3], mode="markers", marker=dict(size=1, color='red'), name='Zn'))
fig.add_trace(go.Scatter3d(x=b2[:,0], y=b2[:,1], z=b2[:,3], mode="markers", marker=dict(size=1, color='blue'), name='Zn'))

fig.update_layout(scene=dict(
                xaxis=dict(title='kx', showgrid=True,range=(0,1)),
                yaxis=dict(title='ky', showgrid=True,range=(0,1)),
                zaxis=dict(title='E (eV)'),),
                title=f'Potential = 0 eV')

fig.show()
```
