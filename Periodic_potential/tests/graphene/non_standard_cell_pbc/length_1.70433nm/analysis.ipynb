```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
directory_path = "./"
def extract_subdirectories(directory_path):
    subdirectories = []
    for foldername, subfolders, filenames in os.walk(directory_path):
        subdirectories.extend(
            subfolder
            for subfolder in subfolders
            if subfolder.startswith("potential_")
        )
    return subdirectories

# subdirectories_list = extract_subdirectories(directory_path)
subdirectories_list = ['potential_0','potential_.1','potential_1']
```

```python
file = 'potential_0/TG_hr_with_atom_coord.dat'
structure = np.loadtxt(file, usecols=(3,5,6,7))
df = pd.DataFrame(structure)
column_name = 0
df = df.drop_duplicates(subset=column_name).reindex()
df
```

```python

x = df[1]
y = df[2]
lattice_period =17.0433
L = lattice_period # should be calculated as the x coord of the first lattice vector
b = L/2 #should be the x-coordinate of the mid point of the ribbon

U0 =  5
colors = []
for i in range(len(x)):
    # if 0 <= x.iloc[i] < b:
    if x.iloc[i] <= b:
        print(x.iloc[i])
        V = U0 / 2.0
        color = 'red'
    else:
        V = -U0 / 2.0
        color = 'black'
    colors.append(color)
plt.figure(figsize=(8,0.5))
plt.scatter(x, y, c=colors, cmap=plt.cm.get_cmap('coolwarm', 2))
plt.show()
```

```python
natoms = 32
for subdirectory in subdirectories_list:
    Hk = np.zeros((natoms,natoms))
    Hdata = np.loadtxt(f'{subdirectory}/TG_hr.dat',skiprows=8,usecols=(3,4,5))
    for i in range(len(Hdata)):
        atom1 = int(Hdata[i,0]-1)
        atom2 = int(Hdata[i,1]-1)
        Hk[atom1,atom2] = Hdata[i,2]
    plt.matshow(Hk)
    plt.title(f'{subdirectory}')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data loading
data = np.loadtxt('./potential_.1/spectrum_unfold_kpath.dat')
x = data[:, 0]  # X values
y = data[:, 1]  # Y values
z = data[:, 5]  # Z values (6th column)

# Define emin and emax
emin = -1.0
emax = 1.0

# Define the colormap
cmap = plt.get_cmap('viridis')

# Determine unique x and y values for reshaping z
unique_x = np.unique(x)
unique_y = np.unique(y)

# Reshape z to match the dimensions of x and y
z_reshaped = z.reshape(len(unique_y), len(unique_x))

# Plotting
fig, ax = plt.subplots(figsize=(19.2, 16.8))

# Create a 2D plot with color mapping
pcm = ax.imshow(z_reshaped, cmap=cmap, origin='lower', extent=[min(unique_x), max(unique_x), min(unique_y), max(unique_y)], aspect='auto', vmin=emin, vmax=emax)

# Set labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Energy (eV)')
ax.set_title('2D Plot with Color Mapping')

# Adding colorbar
cbar = plt.colorbar(pcm, ax=ax)
cbar.set_label('Z Values')

# Adding arrows (assuming the positions are the same as in the GNUplot script)
ax.arrow(0.98308, emin, 0, emax - emin, head_width=0.05, head_length=0.1, fc='k', ec='k', lw=3)
ax.arrow(1.63333, emin, 0, emax - emin, head_width=0.05, head_length=0.1, fc='k', ec='k', lw=3)

# Set y-axis limits
ax.set_ylim(min(unique_y), max(unique_y))

# Set xticks
ax.set_xticks([0.00000, 0.98308, 1.63333, 2.91040])
ax.set_xticklabels(['G', 'K', 'M', 'G'])

# Save the plot
# plt.savefig('spectrum_unfold_kpath.png')
plt.show()
```
