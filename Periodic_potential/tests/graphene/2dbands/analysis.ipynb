```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def split_data_by_bands(filename):
    bands = []
    current_band_data = []

    with open(filename, 'r') as file:
        next(file)

        for line in file:
            if line.strip():  # Check if the line is not empty
                data = line.split()
                current_band_data.append([float(val) for val in data])
            else:
                bands.append(current_band_data)
                current_band_data = []

        if current_band_data:
            bands.append(current_band_data)

    return np.array(bands)
```

```python
bands = np.loadtxt("1by1_nonstandard/potential_0/bulkek.dat")
klen = bands[:,3]
energy = bands[:,4]
plt.plot(klen,energy,".")
plt.ylabel("E (eV)")
plt.ylim(-10,10)
plt.xlabel("kx")
plt.legend()
```
