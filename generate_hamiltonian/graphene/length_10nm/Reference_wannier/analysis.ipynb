```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def split_data_by_bands(filename):
    bands = []
    current_band_data = []

    with open(filename, 'r') as file:
        next(file)

        for line in file:
            if line.strip():  # Check if the line is not empty
                data = line.split()
                current_band_data.append([float(val) for val in data])
            else:
                bands.append(current_band_data)
                current_band_data = []

        if current_band_data:
            bands.append(current_band_data)

    return np.array(bands)
```

```python
bands = split_data_by_bands("sparse/potential_0/bulkek.dat")
vb =  bands[19][:,4]
cb1 = bands[20][:,4]
cb2 = bands[21][:,4]
kx = bands[19][:,0]
plt.plot(kx,vb,label=f"V = 0 eV")
plt.plot(kx,cb1)
plt.ylabel("E (eV)")
plt.xlabel("kx")
plt.legend()
```
