```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def split_data_by_bands(filename):
    bands = []
    current_band_data = []

    with open(filename, 'r') as file:
        next(file)

        for line in file:
            if line.strip():  # Check if the line is not empty
                data = line.split()
                current_band_data.append([float(val) for val in data])
            else:
                bands.append(current_band_data)
                current_band_data = []

        if current_band_data:
            bands.append(current_band_data)

    return np.array(bands)
```

```python
wt_filename = f"wt/BLG/rep_3x3/bulkek.dat"
wt_bands = split_data_by_bands(wt_filename)
gen_filename = f"genham/BLG/rep_3x3/nn_1/null/bulkek.dat"
gen_bands = split_data_by_bands(gen_filename)
```
