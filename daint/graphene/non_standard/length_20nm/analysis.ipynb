```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def split_data_by_bands(filename):
    bands = []
    current_band_data = []

    with open(filename, 'r') as file:
        next(file)

        for line in file:
            if line.strip():  # Check if the line is not empty
                data = line.split()
                current_band_data.append([float(val) for val in data])
            else:
                bands.append(current_band_data)
                current_band_data = []

        if current_band_data:
            bands.append(current_band_data)

    return np.array(bands)
```

```python
potentials = ["0", ".622"]
xtics =  [    0.00000,    0.85137,    1.27707,    1.29243,    2.56950,    2.58486,    3.86202]
xlabels = ["G","P","X","W","Y","G","W"]
for potential in potentials:

    bands = split_data_by_bands(f"kronig/potential_{potential}/bulkek.dat")
    vb =  bands[95][:,4]
    cb1 = bands[96][:,4]
    klen = bands[95][:,3]
    plt.plot(klen,vb,'.')
    plt.plot(klen,cb1,'.',label=f"V = {potential} eV")
    # plt.ylim(-0.04,0.04)
    # plt.xlim(0.775,0.925)
    plt.ylabel("E (eV)")
    plt.xlabel("k")
    plt.axhline(0, color='gray', linewidth=1.5)
    # plt.xticks(labels=xlabels,ticks=xtics)
    plt.legend()
    plt.show()
```
